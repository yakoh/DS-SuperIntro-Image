```python
from PIL import Image
import numpy as np
png=Image.open('/content/drive/MyDrive/'+'Colab/test1.png')
jpg=Image.open('/content/drive/MyDrive/'+'Colab/test1.jpg')

png_array=np.array(png)
jpg_array=np.array(jpg)
png_array16=png_array.astype(np.int16)
jpg_array16=jpg_array.astype(np.int16)
diff=abs(png_array16-jpg_array16)
diff_normal=255*diff/(diff.max())
diff_array=diff_normal.astype(np.uint8)
diff_img=Image.fromarray(diff_array)
diff_img.show()
```
